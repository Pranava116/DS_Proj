```python
"""This is an upgraded version of the previous version leveraging q-learning algo"""

import gym
import numpy as np
import tensorflow as tf
import random
random.seed(0)
tf.compat.v1.disable_eager_execution()
tf.random.set_seed(0)
np.random.seed(0)
num_episodes = 4000 #we can increase the accuracy of the average by increasing the number of episodes
discount = 0.99
a = 0.15 #learning rate
exploration_probability	=lambda	episode:50.	/(episode	+	10)

report_interval	=	500
# 
report = (
    "100-ep Average: %.2f | "
    "Best 100-ep Average: %.2f | "
    "Overall Average: %.2f | "
    "Episode: %d"
)

def one_hot(i, n) -> np.array:
     return	np.identity(n)[i].reshape((1,	-1))
    
def print_vals(rewards, episode):
    print(report % (
    np.mean(rewards[-10:]),
    max([np.mean(rewards[i:i+100]) for i in range(len(rewards) - 100)]),
    np.mean(rewards),
    episode))

def main():
    env = gym.make('FrozenLake-v1')
    env.seed(0)
    rewards = []
    n_obs, n_act = env.observation_space.n, env.action_space.n
    obs_curr_time_ph = tf.compat.v1.placeholder(shape = [1, n_obs], dtype=tf.float32)
    obs_next_time_ph = tf.compat.v1.placeholder(shape = [1, n_obs], dtype = tf.float32)
    act_ph = tf.compat.v1.placeholder(shape=(), dtype = tf.int32)
    reward_ph = tf.compat.v1.placeholder(shape=(), dtype = tf.float32)
    q_target_ph = tf.compat.v1.placeholder(shape= [1,n_act], dtype = tf.float32)
    W = tf.Variable(tf.random.uniform([n_obs,n_act],0,	0.01))
    q_current = tf.matmul(obs_curr_time_ph,	W)
    q_target = tf.matmul(obs_next_time_ph,	W)
    q_target_max = tf.reduce_max(q_target_ph,	axis=1)
    q_target_sa	= reward_ph + discount * q_target_max
    q_current_sa =q_current[0,	act_ph]
    error =	tf.reduce_sum(tf.square(q_target_sa	-q_current_sa))
    pred_act_ph	= tf.argmax(q_current,	1)
    trainer = tf.compat.v1.train.GradientDescentOptimizer(a, use_locking=False, name='GradientDescent')
    update_model = trainer.minimize(error)
    Q_tab = np.zeros((env.observation_space.n, env.action_space.n))
    with tf.compat.v1.Session() as session:
        session.run(tf.compat.v1.global_variables_initializer())
        for episode in range(1, num_episodes+1):
            obs_curr_time = env.reset()
            episode_reward = 0
            while True:
                obs_curr_time_oh = one_hot(obs_curr_time, n_obs)
                action = session.run(pred_act_ph, feed_dict={obs_curr_time_ph: obs_curr_time_oh})[0]
                if np.random.rand(1) < exploration_probability(episode):
                    action = env.action_space.sample()
                env.render()
                obs_next_time,	reward,	done,_=env.step(action)
                obs_next_time_oh =	one_hot(obs_next_time,	n_obs)
                q_target_val	=	session.run(q_target,	feed_dict={obs_next_time_ph:	obs_next_time_oh})
                session.run(update_model,	feed_dict={obs_curr_time_ph:obs_curr_time_oh,reward_ph:	reward,q_target_ph:	q_target_val,act_ph:action})
                
                Q_targ = reward+discount*np.max(Q_tab[obs_next_time, :]) #bellman equation 
                Q_tab[obs_curr_time, action] = (1-a)*Q_tab[obs_curr_time, action]+a*Q_targ
                episode_reward += reward
                obs_curr_time = obs_next_time
                if done:
                    rewards.append(episode_reward)
                    break
    print_vals(rewards, episode)

if __name__ == '__main__':
    main()
```
